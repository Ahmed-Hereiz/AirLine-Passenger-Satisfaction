# Datasets :

```python
X_train = pd.read_csv('Datasets/xtrain_data.csv',index_col='Unnamed: 0')
y_train = pd.read_csv('Datasets/ytrain_data.csv',index_col='Unnamed: 0')
X_test = pd.read_csv('Datasets/xtest_data.csv',index_col='Unnamed: 0')
y_test = pd.read_csv('Datasets/ytest_data.csv',index_col='Unnamed: 0')

assert X_train.shape[0] == y_train.shape[0]
assert X_test.shape[0] == y_test.shape[0]
assert X_train.shape[1] == X_test.shape[1]
assert y_train.shape[1] == y_test.shape[1]
```

### Ok these datasets is unclean and they are like the data that will come from the source... I will use the pipelines on them then I will train some models on them to see the best cleaning pipeline and use it

### These models will help me to choose the best pipeline

### Since that I am going to reuse this on 4 pipelines so I will make function for that
### I made a function that will ask for the models you want to test and names to dispaly for them and x,y data and it will do the cross validation for you....

### function is named as `compute_pipeline()` and is found in `Pipelines.py`, let's use it :

```python
models = []

models.append(RandomForestClassifier())
models.append(BaggingClassifier())
models.append(AdaBoostClassifier())
models.append(ExtraTreesClassifier())
models.append(XGBClassifier())
models.append(LGBMClassifier())

model_name = ['Random Forest','Bagging Classifier', 'Adaboost', 'Extra trees', 'XGB Classifier', 'LGBM Classifier']
```

### I will test these models and there accuracy on many metrics (using my custom function).... these code takes much time to run :(

<br><br><br><br><br><br><br><br><br><br>

<h1><Font color='#3495eb' size=30>Pipeline 1 :</Font></h1>

```python
X_train1, y_train1, full_pipeline, enc = Pipelines.Full_pipeline1(X_train,y_train)

pipeline1_data = Pipelines.compute_pipeline(X_train1,y_train1,models,model_name,1)
pipeline1_data
```

```python
X_train2, y_train2, full_pipeline, enc = Pipelines.Full_pipeline2(X_train,y_train)

pipeline2_data = Pipelines.compute_pipeline(X_train2,y_train2,models,model_name,2)
pipeline2_data
```

<h1><Font color='#3495eb' size=30>Pipeline 3 :</Font></h1>

```python
X_train3, y_train3, full_pipeline, enc = Pipelines.Full_pipeline3(X_train,y_train)

pipeline3_data = Pipelines.compute_pipeline(X_train3,y_train3,models,model_name,3)
pipeline3_data
```

<h1><Font color='#3495eb' size=30>Pipeline 4 :</Font></h1>

```python
X_train4, y_train4, full_pipeline, enc = Pipelines.Full_pipeline4(X_train,y_train)

pipeline4_data = Pipelines.compute_pipeline(X_train4,y_train4,models,model_name,4)
pipeline4_data
```

## Ok last thing... let's see our piplines and choose one of them :

```python
d = {'avg_time_of_models' : [pipeline1_data.loc['Time'].mean(),pipeline2_data.loc['Time'].mean(),
                      pipeline3_data.loc['Time'].mean(),pipeline4_data.loc['Time'].mean()],
'avg_f1_score_of_models' : [pipeline1_data.loc['F1_score'].mean(),pipeline2_data.loc['F1_score'].mean(),
                          pipeline3_data.loc['F1_score'].mean(),pipeline4_data.loc['F1_score'].mean()],
'avg_acc_of_models' : [pipeline1_data.loc['acurracy'].mean(),pipeline2_data.loc['acurracy'].mean(),
                     pipeline3_data.loc['acurracy'].mean(),pipeline4_data.loc['acurracy'].mean()],
'avg_recall_of_models' : [pipeline1_data.loc['recall'].mean(),pipeline2_data.loc['recall'].mean(),
                        pipeline3_data.loc['recall'].mean(),pipeline4_data.loc['recall'].mean()],
'avg_roc_auc_of_models' : [pipeline1_data.loc['roc_auc'].mean(),pipeline2_data.loc['roc_auc'].mean(),
                         pipeline3_data.loc['roc_auc'].mean(),pipeline4_data.loc['roc_auc'].mean()]}

all_pipelines = pd.DataFrame(d,index=['Pipeline1','Pipeline2','Pipeline3','Pipeline4'])
all_pipelines
```
